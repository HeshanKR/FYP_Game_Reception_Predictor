```python
# Check the section column distribution
print("=== PLATFORM DISTRIBUTION ===")
print(meta['section'].value_counts().head(20))

# Filter to PC only
meta_pc = meta[meta['section'] == 'PC']
print(f"\nPC games only: {len(meta_pc)} rows")
print(f"Bombing cases in PC games (gap > 40): {(meta_pc['score_gap'] > 40).sum()}")

# Re-check overlap with Steam after filtering to PC
meta_pc_names = set(meta_pc['title'].str.lower().str.strip())
overlap_pc = meta_pc_names.intersection(steam_names)
print(f"\nOverlap after PC filter: {len(overlap_pc)} games")

matched_pc = meta_pc[meta_pc['title'].str.lower().str.strip().isin(overlap_pc)]
print(f"Bombing cases in overlap after PC filter: {(matched_pc['score_gap'] > 40).sum()}")
```

```python
# Install rapidfuzz for fuzzy matching
!pip install rapidfuzz

from rapidfuzz import process, fuzz

# Take the unmatched Metacritic games and try fuzzy matching
meta_unmatched = meta_pc[~meta_pc['title'].str.lower().str.strip().isin(overlap_pc)]
steam_name_list = list(steam_names)

# Try fuzzy matching for a sample of unmatched games
print("=== FUZZY MATCH SAMPLES ===")
for title in meta_unmatched['title'].head(20):
    match = process.extractOne(title.lower(), steam_name_list, scorer=fuzz.ratio)
    if match and match[1] > 80:  # 80% similarity threshold
        print(f"'{title}' → '{match[0]}' (score: {match[1]})")
```

```python
# Test different thresholds on the FULL matched set
print("=== BOMBING CASES AT DIFFERENT THRESHOLDS ===")
for threshold in [20, 25, 30, 35, 40]:
    count = (matched_pc['score_gap'] > threshold).sum()
    pct = count / len(matched_pc) * 100
    print(f"Gap > {threshold}: {count} bombing cases ({pct:.2f}%)")
```

```python
import ast

def safe_parse_list(val):
    try:
        result = ast.literal_eval(val)
        return result if isinstance(result, list) else []
    except:
        return []

def safe_parse_dict(val):
    try:
        result = ast.literal_eval(val)
        return result if isinstance(result, dict) else {}
    except:
        return {}

steam_valid['genres_parsed']     = steam_valid['genres'].apply(safe_parse_list)
steam_valid['categories_parsed'] = steam_valid['categories'].apply(safe_parse_list)
steam_valid['tags_parsed']       = steam_valid['tags'].apply(safe_parse_dict)
steam_valid['publishers_parsed'] = steam_valid['publishers'].apply(safe_parse_list)
steam_valid['developers_parsed'] = steam_valid['developers'].apply(safe_parse_list)

# Verify parsing worked
print(steam_valid['genres_parsed'].head(3))
print(steam_valid['tags_parsed'].head(2))
```

```python
# Get all unique genres
from collections import Counter

all_genres = Counter()
for g in steam_valid['genres_parsed']:
    all_genres.update(g)

print("Top 15 genres:")
print(all_genres.most_common(15))

# One-hot encode top genres
top_genres = [g for g, _ in all_genres.most_common(10)]
for genre in top_genres:
    col_name = f"genre_{genre.lower().replace(' ', '_')}"
    steam_valid[col_name] = steam_valid['genres_parsed'].apply(
        lambda x: 1 if genre in x else 0
    )

print(f"\nGenre columns created: {[c for c in steam_valid.columns if c.startswith('genre_')]}")
```

```python
genre_cols = [c for c in steam_valid.columns if c.startswith('genre_')]
feature_cols = ['price', 'dlc_count', 'required_age'] + genre_cols

X_v2 = steam_valid[feature_cols].fillna(0)
y_v2 = steam_valid['well_received']

rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
f1_v2 = cross_val_score(rf, X_v2, y_v2, cv=skf, scoring='f1')

print(f"With genres added:")
print(f"F1 scores: {f1_v2.round(3)}")
print(f"Mean F1:   {f1_v2.mean():.3f}  (was 0.618)")
print(f"Std F1:    {f1_v2.std():.3f}  (was 0.017)")
```

```python
# Step 1 — Find most common tags across all games
from collections import Counter

tag_counter = Counter()
for tag_dict in steam_valid['tags_parsed']:
    if isinstance(tag_dict, dict):
        tag_counter.update(tag_dict.keys())

print("=== TOP 30 MOST COMMON TAGS ===")
for tag, count in tag_counter.most_common(30):
    print(f"  {tag}: appears in {count} games")
```

```python
# Step 2 — One-hot encode top 20 tags
top_tags = [tag for tag, _ in tag_counter.most_common(20)]
print("Tags being encoded:", top_tags)

for tag in top_tags:
    col = f"tag_{tag.lower().replace(' ', '_').replace(':', '').replace('-', '_')}"
    steam_valid[col] = steam_valid['tags_parsed'].apply(
        lambda x: 1 if isinstance(x, dict) and tag in x else 0
    )

tag_cols = [c for c in steam_valid.columns if c.startswith('tag_')]
print(f"\n{len(tag_cols)} tag columns created")
```

```python
# Step 3 — Retrain with genres + tags
genre_cols = [c for c in steam_valid.columns if c.startswith('genre_')]
feature_cols = ['price', 'dlc_count', 'required_age'] + genre_cols + tag_cols

X_v3 = steam_valid[feature_cols].fillna(0)
y_v3 = steam_valid['well_received']

rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
f1_v3 = cross_val_score(rf, X_v3, y_v3, cv=skf, scoring='f1')

print(f"With genres + tags:")
print(f"F1 scores: {f1_v3.round(3)}")
print(f"Mean F1:   {f1_v3.mean():.3f}  (was 0.745)")
print(f"Std F1:    {f1_v3.std():.3f}  (was 0.005)")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
import pandas as pd

# Final feature set
genre_cols = [c for c in steam_valid.columns if c.startswith('genre_')]
tag_cols   = [c for c in steam_valid.columns if c.startswith('tag_')]
feature_cols = ['price', 'dlc_count', 'required_age'] + genre_cols + tag_cols

X_final = steam_valid[feature_cols].fillna(0)
y_final = steam_valid['well_received']

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

models = {
    'Logistic Regression': LogisticRegression(
        class_weight='balanced', max_iter=1000, random_state=42),
    'Random Forest':       RandomForestClassifier(
        n_estimators=100, class_weight='balanced', random_state=42),
    'XGBoost':             XGBClassifier(
        n_estimators=100, scale_pos_weight=len(y_final[y_final==0])/len(y_final[y_final==1]),
        random_state=42, eval_metric='logloss', verbosity=0),
    'CatBoost':            CatBoostClassifier(
        iterations=100, auto_class_weights='Balanced',
        random_seed=42, verbose=0)
}

results = []
for name, model in models.items():
    f1    = cross_val_score(model, X_final, y_final, cv=skf, scoring='f1')
    acc   = cross_val_score(model, X_final, y_final, cv=skf, scoring='accuracy')
    roc   = cross_val_score(model, X_final, y_final, cv=skf, scoring='roc_auc')
    results.append({
        'Model': name,
        'Mean F1':       round(f1.mean(),  3),
        'Std F1':        round(f1.std(),   3),
        'Mean Accuracy': round(acc.mean(), 3),
        'Mean ROC-AUC':  round(roc.mean(), 3)
    })
    print(f"{name}: F1={f1.mean():.3f} ± {f1.std():.3f}")

results_df = pd.DataFrame(results)
print("\n=== FULL COMPARISON TABLE ===")
print(results_df.to_string(index=False))
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

print("=== IMBALANCE STRATEGY COMPARISON (Random Forest) ===\n")

strategies = {
    'No adjustment':     RandomForestClassifier(n_estimators=100, random_state=42),
    'Class weights':     RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42),
    'SMOTE':             ImbPipeline([
                            ('smote', SMOTE(random_state=42)),
                            ('model', RandomForestClassifier(n_estimators=100, random_state=42))
                         ])
}

for name, model in strategies.items():
    f1  = cross_val_score(model, X_final, y_final, cv=skf, scoring='f1')
    print(f"{name:20s}: F1 = {f1.mean():.3f} ± {f1.std():.3f}")
```
